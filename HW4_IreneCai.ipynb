```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score  # 添加了 r2_score

# 加载数据
train_data = pd.read_csv("fill_rep_data.csv")
test_data = pd.read_csv("fill_rep_val.csv")

# 使用 `get_dummies` 并对齐列
data_encoded = pd.get_dummies(train_data, columns=['States'])
data_encoded_test = pd.get_dummies(test_data, columns=['States'])

# 确保测试集具有训练集的所有特征
data_encoded_test = data_encoded_test.reindex(columns=data_encoded.columns, fill_value=0)

# 筛选没有任何数据缺失的行
complete_data = data_encoded.dropna()
complete_data_test = data_encoded_test.dropna()

# 定义特征和目标变量
features = complete_data.columns.drop(['uti_card_50plus_pct'])

# 训练模型
X_train_uti = complete_data[features]
y_train_uti = complete_data['uti_card_50plus_pct']
model_uti = LinearRegression()
model_uti.fit(X_train_uti, y_train_uti)

# 定义测试集的特征和真实值
X_test_uti = complete_data_test[features]
y_test_uti = complete_data_test['uti_card_50plus_pct']

# 进行预测
predicted_uti = model_uti.predict(X_test_uti)

# 计算MSE和R^2
mse_uti = mean_squared_error(y_test_uti, predicted_uti)
r2_uti = r2_score(y_test_uti, predicted_uti)

# 打印评估结果
print("MSE for uti_card_50plus_pct prediction:", mse_uti)
print("R^2 for uti_card_50plus_pct prediction:", r2_uti)

# 使用模型预测缺失的值
missing_data = data_encoded[data_encoded.isnull().any(axis=1)]
missing_data_val = data_encoded_test[data_encoded_test.isnull().any(axis=1)]
missing_data_clean = missing_data.dropna(subset=features)
missing_data_clean_val = missing_data_val.dropna(subset=features)

missing_data_uti_pred = model_uti.predict(missing_data_clean[features])
missing_data_uti_pred_val = model_uti.predict(missing_data_clean_val[features])
data_encoded.loc[missing_data_clean.index, 'uti_card_50plus_pct'] = missing_data_uti_pred
data_encoded_test.loc[missing_data_clean_val.index, 'uti_card_50plus_pct'] = missing_data_uti_pred_val

# 打印填充后的数据，检查是否还有NaN
print(data_encoded.loc[data_encoded['uti_card_50plus_pct'].isnull() | data_encoded['rep_income'].isnull()])
# 打印填充后的数据，检查是否还有NaN
print(data_encoded_test.loc[data_encoded_test['uti_card_50plus_pct'].isnull() | data_encoded_test['rep_income'].isnull()])

# 将处理后的数据保存到CSV文件中
data_encoded.to_csv("fill_all_data.csv", index=False)
data_encoded_test.to_csv("fill_all_data_val.csv", index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# 加载数据
data = pd.read_csv("fill_all_data.csv")
val_data = pd.read_csv("fill_all_data_val.csv")

# 选择特征和目标变量
# 确定特征和目标
X_train = data.drop(['Default_ind'], axis=1)
X_test = val_data.drop(['Default_ind'], axis=1)

y_train = data['Default_ind']
y_test = val_data['Default_ind']


# 拟合逻辑回归模型
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# 预测测试集结果
y_pred = model.predict(X_test)
y_pred_proba = model.coef_[0]

# 计算准确率和ROC AUC
accuracy = accuracy_score(y_test, y_pred)
auc_score = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print("Accuracy:", accuracy)
print("ROC AUC Score:", auc_score)

# 绘制ROC曲线
fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
plt.plot(fpr, tpr, label="AUC = {:.3f}".format(auc_score))
plt.plot([0, 1], [0, 1], 'r--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# 显示混淆矩阵
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

**第三个改进模型


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

# 加载数据
train_data = pd.read_csv("fill_all_data.csv")
test_data = pd.read_csv("fill_all_data_val.csv")

# 准备特征和目标变量
X_train = train_data.drop(['Default_ind'], axis=1)
y_train = train_data['Default_ind']
X_test = test_data.drop(['Default_ind'], axis=1)
y_test = test_data['Default_ind']

# 数据标准化
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 构建神经网络模型
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # 输出层使用sigmoid激活函数，适用于二分类问题

# 编译模型
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 训练模型
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# 预测测试集结果
y_pred_proba = model.predict(X_test).flatten()  # 获取预测概率
y_pred = (y_pred_proba > 0.5).astype(int)  # 将概率转换为类别

# 评估模型
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("ROC AUC Score:", roc_auc)
print("Confusion Matrix:\n", conf_matrix)

# 绘制ROC曲线
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# 筛选出模型预测为违约的贷款
predicted_defaults = test_data[y_pred == 1]

# 计算总坏账
total_bad_debt = predicted_defaults['tot_credit_debt'].sum()
ave_bad_debt = predicted_defaults['avg_card_debt'].sum()

# 打印总坏账
print("Total Bad Debt Amount:", total_bad_debt)
print("Average Bad Debt Amount:", ave_bad_debt)
```

```python
import matplotlib.pyplot as plt



# 将预测结果添加到测试数据集中，我们用一个新列 "predicted_default" 来存储
test_data['predicted_default'] = y_pred

# 加载数据（假设已经包含预测结果）
data = test_data

# 划分不同的情况
pred_loan_act_bad = data[(data['predicted_default'] == 0) & (data['Default_ind'] == 1)]  # 预测放款，实际坏账
pred_loan_act_good = data[(data['predicted_default'] == 0) & (data['Default_ind'] == 0)]  # 预测放款，实际好账
pred_bad_act_good = data[(data['predicted_default'] == 1) & (data['Default_ind'] == 0)]  # 预测坏账，实际好账
pred_bad_act_bad = data[(data['predicted_default'] == 1) & (data['Default_ind'] == 1)]  # 预测坏账，实际坏账

# 绘制散点图
plt.figure(figsize=(12, 8))
plt.scatter(pred_loan_act_good.index, pred_loan_act_good['avg_card_debt'], color='blue', label='Predicted Loan, Actual Good', alpha=0.5)
plt.scatter(pred_loan_act_bad.index, pred_loan_act_bad['avg_card_debt'], color='red', label='Predicted Loan, Actual Bad', alpha=0.5)
plt.scatter(pred_bad_act_good.index, pred_bad_act_good['avg_card_debt'], color='green', label='Predicted Bad, Actual Good', alpha=0.5)
plt.scatter(pred_bad_act_bad.index, pred_bad_act_bad['avg_card_debt'], color='yellow', label='Predicted Bad, Actual Bad', alpha=0.5)

# 添加图形元素
plt.title('Average Credit Debt by Predicted and Actual Default Status')
plt.xlabel('Index')
plt.ylabel('Average Credit Debt')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```
